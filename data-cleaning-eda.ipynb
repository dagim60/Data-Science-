# Data cleaning

```python
# Rename columns to all lowercase and no spaces
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df.columns
```

```python
try:
    df.drop(["owner", "certification_address", "certification_contact","altitude",
              "status","mill",
             "lot_number","ico_number"],
            axis=1, inplace=True)
except KeyError:
    pass
```

```python
#making ID as DF index and removing Unnamed columns
df.set_index("id",inplace = True)
df.drop("unnamed:_0",axis=1,inplace=True)
```

```python
df.head()
```

```python
# removing chinese character from "In-company Partner" and "Producer"
import re 
#to extract only english characters
def extract_eng_char(s):
    s = str(s)
    s = re.sub(r'[^a-zA-Z0-9\s]', '', s).strip()
    return s

columns_with_noneng = ["producer","in-country_partner","certification_body",
                       "farm_name","region"]
for col in columns_with_noneng:
    df[col] = df[col].apply(extract_eng_char)
    #Set unknown as producer for empty strings
    df[col] = df[col].replace("","unknown")
```

```python
df['color'].value_counts()
```

```python
#As we can see from the above results there are some typos in the 'Color' features. Therefore, I will correct these typos
# Correcting misspelled data
df['color'] = df['color'].str.replace('yellow green','yellow-green')
df['color'] = df['color'].str.replace('yellow- green','yellow-green')
df['color'] = df['color'].str.replace('yello-green','yellow-green')

df['color'].value_counts()
```

Overall, Flavor, Aftertaste, Balance, Acidity, Aroma, Body: These variables have strong positive correlations with 'Total Cup Points'. This means that as these variables increase, the 'Total Cup Points' also tend to increase. For example, a higher 'Overall' rating or 'Flavor' rating is associated with a higher 'Total Cup Points' score.

Processing Method_Pulped natural / honey, Altitude, Coffee Age, Processing Method_Washed / Wet: These variables have weak positive correlations with 'Total Cup Points'. This suggests that these variables might have a slight positive effect on 'Total Cup Points', but the relationship is not as strong as the previous variables.

Moisture Percentage, Category One Defects: These variables have weak negative correlations with 'Total Cup Points'. This means that as these variables increase, the 'Total Cup Points' tend to decrease slightly. However, the correlation is weak, so these variables might not have a significant impact on 'Total Cup Points'.

Processing Method_Natural / Dry, Category Two Defects, Quakers: These variables have moderate negative correlations with 'Total Cup Points'. This suggests that as these variables increase, the 'Total Cup Points' tend to decrease. For example, a higher number of 'Category Two Defects' or 'Quakers' is associated with a lower 'Total Cup Points' score.

# display the relationship between flavor and aftertaste scores for the coffee samples

```python

# Plotting a scatter plot of flavor and aftertaste scores
plt.figure(figsize=(8, 6))  # Set the figure size

# Plotting the scatter plot
plt.scatter(df1['flavor'], df1['aftertaste'], alpha=0.7 )

# Adding labels and title
plt.xlabel('flavor')
plt.ylabel('aftertaste')
plt.title('Scatter Plot of Flavor and Aftertaste Scores')

# Displaying the plot
plt.show()
```


depending on previous plot there is a positive correlation between flavor and aftertaste scores. This means that as the flavor score increases, the aftertaste score also tends to be higher within defferent coffee Samples
